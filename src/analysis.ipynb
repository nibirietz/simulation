```python
import subprocess

repeats = 20

for run_num in range(1, repeats + 1):
    result = subprocess.run(
        ["python", "main.py"],
        text=True
    )
```

```python
h_deaths = []
p_deaths = []

for i in range(0, repeats):
    h_deaths.append(0)
    p_deaths.append(0)

    with open(f"logs/animals_{i + 1}.log") as file:
        for line in file:
            line = line.split(",")
            if line[2].strip() == "dead":
                if line[1] == "Herbivore":
                    h_deaths[i] += 1
                else:
                    p_deaths[i] += 1
```

```python
sum(h_deaths) / repeats, sum(p_deaths) / repeats
```
